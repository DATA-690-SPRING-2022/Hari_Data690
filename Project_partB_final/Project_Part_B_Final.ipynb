```python
#Importing libraries

import pandas as pd
import plotly
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
```

```python
pio.templates
```

```python
pio.templates.default='plotly_dark'
```

```python
#Reading data

df = pd.read_csv('https://raw.githubusercontent.com/harikrishna0007/data_609_Hari/main/Individual_project_Part_B/Data.csv')
df.head(10)
```

```python
#Dropping unwanted rows 

df_new = df.drop(columns=['Unnamed: 0'])
df_new.head()
```

```python
#Taking only particular data of 2020 year

df_2020 = df_new.query("Year == 2020")
df_2020.sample(10)
```

```python
#Filtering only 2020 gdp data

df_2020_gdp = df_2020.query("indicator in['CM.MKT.TRAD.GD.ZS']")
df_2020_gdp
```

```python
#Plotting bar graph of Stocks Traded- Total Value(% of GDP) in the year 2020


fig = px.bar(
    data_frame=df_2020_gdp,
    title="Stocks Traded- Total Value(% of GDP) in the year 2020",
    labels={"value":"Total Value(% of GDP)"},
    x="Country Name",
    y="value",
    color="Country Name"
)

fig.update_xaxes(categoryorder="total descending")


fig.show()
```

```python
#Stocks Traded Turnover ratio of domestic shares in the year 2020

df_2020_ds = df_2020.query("indicator in['CM.MKT.TRNR']")
df_2020_ds
```

```python
#Plotting the bar graph for the year 2020

fig = px.bar(
    data_frame=df_2020_ds,
    title="Stocks Traded Turnover ratio of domestic shares in the year 2020",
    labels={"value":"Domestic Share of Stocks", "Year":"YEAR"},
    x="Country Name",
    y="value",
    color="Country Code"
)

fig.update_xaxes(categoryorder="total descending")


fig.show()
```

```python
#Stocks Traded Total Value(Current US Dollars) in the year 2020

df_2020_CR = df_2020.query("indicator in['CM.MKT.TRAD.CD']")
df_2020_CR
```

```python
#Plotting pie graph for Stocks Traded Total Value(Current US Dollars) in the year 2020

fig = px.pie(
    df_2020_CR,
    values='value', 
    names='Country Name',
    title='Stocks Traded Total Value(Current US Dollars) in the year 2020',
    hover_name='Country Code',
)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```
